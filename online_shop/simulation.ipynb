```python
from faker import Faker
import random # random numbers
import pandas as pd
import faker_commerce
```

```python
fake = Faker("de_DE")
fake.add_provider(faker_commerce.Provider)

# fake Users

rows = []
n_users = 10000

for _ in range(n_users): # repeat something n_users-times without running index variable 
    first_name = fake.first_name()
    last_name = fake.last_name()
    user_id = fake.unique.uuid4()
    user = {
        "id": user_id,
        "name": first_name + " " + last_name 
    }
    rows.append(user)

df = pd.DataFrame(rows)
df
```

```python
rows = []
n_products = 500

for _ in range(n_products):
    product = {
        "id": fake.unique.uuid4(),
        "name": fake.ecommerce_name(),
        "price": float(random.randint(10, 1000)) + 0.99, 
        "stock": random.randint(0,40)
    }
    rows.append(product)
df_product = pd.DataFrame(rows)
df_product
```

```python
df_product_m = df_product.copy() # in pandas this is deep copy 

# cat 1 -> costs more then 600
# cat 2 -> costs 300 - 600
# cat 3 -> costs 200 - 300
# cat 4 -> costs < 200

def price_category(price):
    if price>600:
        return 1
    elif price>300:
        return 2
    elif price > 200:
        return 3
    else:
        return 4

df_product_m["price_category"] = df_product_m["price"].apply(lambda p: price_category(p))
df_product_m
```

```python
# add column out_of_stock
df_product_m["out_of_stock"] = df_product_m["stock"] == 0
df_product_m["out_of_stock"].sum()
```

```python
df_product_m[df_product_m["price_category"] == 1].stock.mean()
```

```python
agg_df = df_product_m.groupby(["price_category"])[["price", "stock"]].agg(
    [
        "mean",
        "std",
        "min",
        "max",
    ]
)

agg_df
```

```python
def filter_category_one(df): return df[df.price_category == 1]
def filter_on_stock(df): return df[df.out_of_stock == False]

# executing the filters in a pipe 

# data transformation pipeline

df_category_one_on_stock = (df_product_m
    .pipe(filter_category_one)
    .pipe(filter_on_stock)
    .sort_values("price", ascending=False)
)
df_category_one_on_stock
```
