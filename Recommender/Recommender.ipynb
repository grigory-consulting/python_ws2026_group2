```python
import pandas as pd
import math

user_data = pd.read_csv("u.data",
                        sep= "\t", 
                        names=["user_id", "movie_id", "rating", "timestamp"],
                        usecols=[0, 1, 2],
                        )


main_dict = {}

for movie in user_data.movie_id:
    movie_subset = user_data[user_data.movie_id == movie] 
    main_dict[movie] = dict(zip(movie_subset.user_id, movie_subset.rating))

```
